```python
import numpy as np
import matplotlib.pyplot as plt

given_bits = input("Bits: ")

bits_list = [int(char) for char in given_bits]
print(bits_list)

#ALTERNATE MARK INVERSION ENCODING
def plot_signal(encoded_signal):
  time = range(len(encoded_signal))
  plt.step(time, encoded_signal, where = 'post')
  plt.xlabel('Time')
  plt.ylabel('Signal')
  plt.title('AML Input')
  plt.grid(True)
  plt.show()

plot_signal(bits_list)

def ami_encoding(given_bits):
  encoded_signal = []
  count = 0
  for bit in given_bits:
    if bit == '1':
      if count%2==0:
        encoded_signal.append(1)
      else:
        encoded_signal.append(-1)
      count+=1
    else:
      encoded_signal.append(0)
  return encoded_signal

encoded = ami_encoding(given_bits)
print(f"Encoded Signal: {encoded}")

def plot_encoded_signal(encoded_data):
  time = range(len(encoded_data))
  plt.step(time, encoded_data, where='post')
  plt.xlabel('Time')
  plt.ylabel('Voltage')
  plt.title('AMI Encoding')
  plt.grid(True)
  plt.show()

plot_encoded_signal(encoded)

def ami_decoding(encoded):
  decoded_data = ''
  for bit in encoded:
    if bit == 1:
      decoded_data += '1'
    elif bit == -1:
      decoded_data += '1'
    else:
      decoded_data += '0'
  return decoded_data

decoded = [int(char) for char in ami_decoding(encoded)]
print(f"Decoded Signal: {decoded}")

def plot_decoded_data(encoded_signal):
  time = range(len(encoded_signal))
  plt.step(time, encoded_signal, where='post')
  plt.xlabel('Time')
  plt.ylabel('Signal')
  plt.title('AMI Decoding')
  plt.grid(True)
  plt.show()

plot_decoded_data(decoded)
```
