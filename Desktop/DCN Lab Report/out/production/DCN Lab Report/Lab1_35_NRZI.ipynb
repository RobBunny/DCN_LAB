```python
import numpy as np
import matplotlib.pyplot as plt

given_bits = '10111001'

bits_list = [int(char) for char in given_bits]
print(bits_list)

#NRZ-I ENCODING
def plot_signal(encoded_signal):
  time = range(len(encoded_signal))
  plt.step(time, encoded_signal, where = 'post')
  plt.xlabel('Time')
  plt.ylabel('Signal')
  plt.title('NRZ-I Input')
  plt.grid(True)
  plt.show()

plot_signal(bits_list)

def nrzi_encoding(given_bits, init_state):
  encoded_signal = []
  prev_state = init_state
  for bit in given_bits:
    if bit == '1':
      if prev_state == 1:
        prev_state = 0
      else:
        prev_state = 1
    encoded_signal.append(prev_state)
  return encoded_signal

encoded = nrzi_encoding(given_bits, 1) #Let initial state be high
print(f"Encoded Signal: {encoded}")

def plot_encoded_signal(encoded_data):
  time = range(len(encoded_data))
  plt.step(time, encoded_data, where='post')
  plt.xlabel('Time')
  plt.ylabel('Voltage')
  plt.title('NRZ-I Encoding')
  plt.grid(True)
  plt.show()

plot_encoded_signal(encoded)

def nrzi_decoding(encoded):
  decoded_data = '1'
  for i in range(1, len(encoded)):
    if encoded[i] == encoded[i-1]:
      decoded_data += '0'
    else:
      decoded_data += '1'
  return decoded_data

decoded = [int(char) for char in nrzi_decoding(encoded)]
print(f"Decoded Signal: {decoded}")

def plot_decoded_data(encoded_signal):
  time = range(len(encoded_signal))
  plt.step(time, encoded_signal, where='post')
  plt.xlabel('Time')
  plt.ylabel('Signal')
  plt.title('NRZ-I Decoding')
  plt.grid(True)
  plt.show()

plot_decoded_data(decoded)
```
