```python
import numpy as np
import matplotlib.pyplot as plt

given_bits = '10111001'

bits_list = [int(char) for char in given_bits]
print(bits_list)

#MANCHESTER ENCODING

def plot_signal(encoded_signal):
  time = range(len(encoded_signal))
  plt.step(time, encoded_signal, where = 'post')
  plt.xlabel('Time')
  plt.ylabel('Signal')
  plt.title('Manchester Input')
  plt.grid(True)
  plt.show()

plot_signal(bits_list)

def mnchstr_encoding(given_bits):
  encoded_signal = []
  for bit in given_bits:
    if bit == '1':
      encoded_signal.append(0)
      encoded_signal.append(1)
    else:
      encoded_signal.append(1)
      encoded_signal.append(0)
  return encoded_signal

encoded = mnchstr_encoding(given_bits)
print(f"Encoded Signal: {encoded}")

def plot_encoded_signal(encoded_data):
  time = np.arange(len(encoded_data))/2
  plt.figure(figsize=(10, 3))
  plt.step(time, encoded_data, where='post')
  plt.xlabel('Time')
  plt.ylabel('Amplitude')
  plt.title('Manchester Encoding')
  plt.grid(True)
  plt.show()

plot_encoded_signal(encoded)

def mnchstr_decoding(encoded):
  decoded_data = ''
  for i in range(0, len(encoded), 2):
    if encoded[i] == 1 and encoded[i+1] == 0:
      decoded_data += '0'
    elif encoded[i] == 0 and encoded[i+1] == 1:
      decoded_data += '1'
  return decoded_data

decoded = [int(char) for char in mnchstr_decoding(encoded)]
print(f"Decoded Signal: {decoded}")

def plot_decoded_data(encoded_signal):
  time = range(len(encoded_signal))
  plt.step(time, encoded_signal, where='post')
  plt.xlabel('Time')
  plt.ylabel('Signal')
  plt.title('Manchester Decoding')
  plt.grid(True)
  plt.show()

plot_decoded_data(decoded)
```
