```python
import numpy as np
import matplotlib.pyplot as plt

given_bits = '10111001'

bits_list = [int(char) for char in given_bits]
print(bits_list)

#NRZ-L ENCODING
def plot_signal(encoded_signal):
  time = range(len(encoded_signal))
  plt.step(time, encoded_signal, where = 'post')
  plt.xlabel('Time')
  plt.ylabel('Signal')
  plt.title('NRZ-L Input')
  plt.grid(True)
  plt.show()

plot_signal(bits_list)

def nrzl_encoding(given_bits):
  encoded_signal = []
  for bit in given_bits:
    if bit == '1':
      encoded_signal.append(1)
    else:
      encoded_signal.append(0)
  return encoded_signal

encoded = nrzl_encoding(given_bits)
print(f"Encoded Signal: {encoded}")

def plot_encoded_signal(encoded_data):
  time = np.linspace(0, len(encoded_data), len(encoded_data)*100)
  signal = np.repeat(encoded_data, 100)

  plt.step(time, signal, where='post')
  plt.xlabel('Time')
  plt.ylabel('Voltage')
  plt.title('NRZ-L Encoding')
  plt.grid(True)
  plt.show()

plot_encoded_signal(encoded)

def nrzl_decoding(encoded):
  decoded_data = ''
  for bit in encoded:
    if bit == 1:
      decoded_data += '1'
    else:
      decoded_data += '0'
  return decoded_data

decoded = [int(char) for char in nrzl_decoding(encoded)]
print(f"Decoded Signal: {decoded}")

def plot_decoded_data(encoded_signal):
  time = range(len(encoded_signal))
  plt.step(time, encoded_signal, where='post')
  plt.xlabel('Time')
  plt.ylabel('Signal')
  plt.title('NRZ-L Decoding')
  plt.grid(True)
  plt.show()

plot_decoded_data(decoded)
```
